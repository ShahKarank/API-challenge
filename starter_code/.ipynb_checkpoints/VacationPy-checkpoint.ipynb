### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(g_key)
locations = cname[["Lat","Lng"]]
humidity = cname["Humidity"]

layout = {
    'width': '1000px',
    'height': '600px'
        }


plt1 = gmaps.figure(map_type='HYBRID', layout=layout)

heat = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 1)

plt1.add_layer(heat)

plt1
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
burl = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

goodW = cname.loc[(cname['Temperature'] >= 70) & (cname['Temperature'] <= 80) & 
                                  (cname['Cloudiness'] == 0)]
hotel = goodW
hotel['Hotel Name']=""

for index, x in hotel.iterrows():
    
    (lat,lng) = hotel.loc[index, ['Lat','Lng']]

    params = {
        "location": f"{lat},{lng}", 
        "radius": "5000",
        "type": "hotel",
        "key": g_key,
    }

    response = requests.get(burl, params=params).json()
    
    results = response['results']

    try:
        print(f"Closest hotel is {results[1]['name']}.")
        
        hotel.loc[index, 'Hotel Name'] = results[1]['name']
     
    except (KeyError, IndexError):
        print("Missing => skipping.")
        
    print(" ")
hotel
```
